```python
import json
import pandas as pd

filename = "customers_generated_100000_seed42.jsonl"
batch = []
with open(filename, 'r') as file:
    for line in file:
        batch.append(json.loads(line))
        if len(batch) == 10000:
            break
```

```python
len(batch)
```

```python
for profile in batch:
    # check to see if there are multiple payment methods in a single profile from the same paymentMethodIssuer
    payment_method_issuers = set()
    for payment_method in profile.get('paymentMethods', []):
        issuer = payment_method.get('paymentMethodIssuer')
        if not issuer:
            print(f"Missing paymentMethodIssuer in customer {profile.get('customer', {}).get('customerEmail', 'unknown')}")
        if issuer:
            if issuer in payment_method_issuers:
                print(f"Duplicate payment method issuer found: {issuer} in customer {profile.get('customer')["customerEmail"]}")
                print(f"issuer: {issuer}")
            else:
                payment_method_issuers.add(issuer)
    
```

```python
batch_df = pd.DataFrame(batch)
```

```python
# get the loggedAt column from transactions
transactions_series = batch_df.transactions
```

```python
logged_at_list = transactions_series.apply(lambda transactions: [trans.get("loggedAt")[:11] for trans in transactions]).explode().tolist()
```

```python
times = set(logged_at_list)
times
times_2025 = {t for t in times if str(t).startswith("2025-08-2")}
times_2025
```
